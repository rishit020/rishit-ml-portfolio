```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)

n = 100
X = np.linspace(-10, 10, n)
y = 3 * X + 5 + np.random.randn(n) * 3


lr = 0.01
epochs = 1000

w = 0.0
b = 0.0


for _ in range(epochs):
    y_pred = w * X + b

    dw = np.mean((y_pred - y) * X)
    db = np.mean(y_pred - y)

    w -= lr * dw
    b -= lr * db


x_line = np.linspace(X.min(), X.max(), 100)
y_line = w * x_line + b


plt.scatter(X, y)
plt.plot(x_line, y_line)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression from Scratch")
plt.show()
```
