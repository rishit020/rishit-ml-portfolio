```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)

n = 200
X = np.random.randn(n, 2)
y = (X[:, 0] + X[:, 1] > 0).astype(int)


def sigmoid(z):
    z = np.clip(z, -500, 500)
    return 1 / (1 + np.exp(-z))


lr = 0.1
epochs = 1000

w = np.zeros(2)
b = 0.0


for _ in range(epochs):
    z = X @ w + b
    p = sigmoid(z)

    dw = (X.T @ (p - y)) / n
    db = np.mean(p - y)

    w -= lr * dw
    b -= lr * db


x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x_vals = np.linspace(x_min, x_max, 100)
y_vals = -(w[0] * x_vals + b) / w[1]


plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1])
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1])
plt.plot(x_vals, y_vals)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Logistic Regression from Scratch")
plt.show()
```

```python
z_vals = X @ w + b
p_vals = sigmoid(z_vals)

sorted_idx = np.argsort(z_vals)
z_sorted = z_vals[sorted_idx]
p_sorted = p_vals[sorted_idx]

plt.figure()
plt.scatter(z_vals, y)
plt.plot(z_sorted, p_sorted)
plt.xlabel("z = wÂ·x + b")
plt.ylabel("Probability")
plt.title("Sigmoid Function Learned by Logistic Regression")
plt.show()
```
